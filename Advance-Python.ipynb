```python
#Python program to identify duplicate student names using sets by converting names to lowercase and removing spaces.
from datetime import datetime
a = set()
b = set()
n = int(input("Enter the number of students: "))
for i in range(n):
    name = input("Enter the name of the student: ")
    clean_name = name.lower().replace(" ", "")
    current_time = datetime.now().strftime("%H:%M:%S")
    if clean_name in a:
        b.add(clean_name)
        print(f"[{current_time}] {clean_name} is FAKE (already exists)")
    else:
        a.add(clean_name)
        print(f"[{current_time}] {clean_name} is REAL (added successfully)")
print("\nReal students:", a)
print("Fake students:", b)
```

```python
#program allows only unique student entries using a set and rejects duplicate entries, storing them in a separate list.
entered_std = set()
rejected_std = []

n = int(input("Enter entry attempts: "))

for i in range(n):
    name = input("Enter student names: ")

    if name in entered_std:
        print(name, "Entry Rejected . Already Entered")
        rejected_std.append(name)
    else:
        print(name, "Entry Allowed")
        entered_std.add(name)

print("\nEvent Tracker Summary")
print("Entered Students:", entered_std)
print("Rejected Students:", rejected_std)
```

```python
# library management system
import datetime

books = [
    ["Python Programming", "Mrs. Sree Lakshmi"],
    ["Data Analysis and Algorithm", "Mrs. Nirupama Dora"],
    ["Operating System", "Mrs. Supriya Sathpathy"],
    ["Computer Architecture", "Mr. Dilip Kumar Behera"],
    ["Introduction to Machine Learning", "Mr. Tirupathi Sahu"],
    ["English", "Mr. B.V.N Ramanamurthy"]
]

issued_books = []

def show_books():
    if not books:
        print("\nNo books are currently available.\n")
        return

    print("\nAvailable Books:")
    for i, book in enumerate(books):
        print(f"{i+1}. {book[0]} by {book[1]}")
    print()

def issue_book():
    show_books()
    if not books:
        return

    try:
        choice = int(input("Choose book number to issue: ")) - 1
        if choice < 0 or choice >= len(books):
            print("Invalid choice.\n")
            return
    except ValueError:
        print("Please enter a valid number.\n")
        return

    user = input("Enter your name: ").title()
    issue_time = datetime.datetime.now()

    issued_books.append([books[choice][0], books[choice][1], user, issue_time])
    print(f"\n'{books[choice][0]}' issued to {user} on {issue_time.strftime('%d-%m-%Y %H:%M:%S')}\n")

    books.pop(choice)


def return_book():
    if not issued_books:
        print("\nNo books are issued right now.\n")
        return

    print("\nIssued Books:")
    for i, book in enumerate(issued_books):
        print(f"{i+1}. {book[0]} (User: {book[2]})")

    try:
        choice = int(input("Choose book number to return: ")) - 1
        if choice < 0 or choice >= len(issued_books):
            print("Invalid choice.\n")
            return
    except ValueError:
        print("Please enter a valid number.\n")
        return

    return_time = datetime.datetime.now()
    issue_time = issued_books[choice][3]
    time_taken = return_time - issue_time

    books.append([issued_books[choice][0], issued_books[choice][1]])

    print("\nBook returned successfully.")
    print(f"Time taken: {time_taken}\n")

    issued_books.pop(choice)


def check_book():
    name = input("\nEnter book name: ").strip()

    for book in books:
        if book[0].lower() == name.lower():
            print("Book is available.\n")
            return

    for book in issued_books:
        if book[0].lower() == name.lower():
            print(f"Book is currently issued to {book[2]}.\n")
            return

    print("Book not found in library.\n")


def user_details():
    if not issued_books:
        print("\nNo issued book records.\n")
        return

    print("\nUser Issue Details:")
    for book, author, user, time in issued_books:
        print(f"Book: {book} | Author: {author} | User: {user} | Issued on: {time.strftime('%d-%m-%Y %H:%M:%S')}")
    print()


def menu():
    while True:
        print("===== Library Management System =====")
        print("1. Show Available Books")
        print("2. Issue Book")
        print("3. Return Book")
        print("4. Check Book Availability")
        print("5. View User Details")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            show_books()
        elif choice == "2":
            issue_book()
        elif choice == "3":
            return_book()
        elif choice == "4":
            check_book()
        elif choice == "5":
            user_details()
        elif choice == "6":
            print("\nThank you for using the Library Management System.")
            break
        else:
            print("Invalid choice. Try again.\n")


menu()
```

```python
#password strength checker
import re

def check_password_strength(password):
    score = 0

    if len(password) >= 8:
        score += 1
    if re.search(r"[A-Z]", password):
        score += 1
    if re.search(r"[a-z]", password):
        score += 1
    if re.search(r"[0-9]", password):
        score += 1
    if re.search(r"[@$!%*?&]", password):
        score += 1

    if score <= 2:
        return " Weak password – easy to guess"
    elif score <= 4:
        return " Medium password – could be stronger"
    else:
        return " Strong password – good job!"

print("Create a strong password")
password = input("Enter your password: ")

print("\nResult:")
print(check_password_strength(password))
```

```python
# online shopping cart

products = {
    "Laptop": 50000,
    "Phone": 20000,
    "Headphones": 2000,
    "Keyboard": 1500
}

cart = {}

def show_products():
    print("\nAvailable Products:")
    for item, price in products.items():
        print(f"{item} - ₹{price}")

def add_to_cart(item):
    if item in products:
        if item in cart:
            cart[item] += 1
        else:
            cart[item] = 1
        print(f"{item} added to cart ")
    else:
        print("Product not found ")

def view_cart():
    if not cart:
        print("\nCart is empty ")
        return

    print("\nYour Cart:")
    total = 0
    for item, qty in cart.items():
        price = products[item] * qty
        total += price
        print(f"{item} x {qty} = ₹{price}")
    print("Total Amount: ₹", total)

while True:
    print("\n1. Show Products")
    print("2. Add to Cart")
    print("3. View Cart")
    print("4. Checkout")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        show_products()

    elif choice == "2":
        item = input("Enter product name: ")
        add_to_cart(item)

    elif choice == "3":
        view_cart()

    elif choice == "4":
        view_cart()
        print("Thank you for shopping! ")
        break

    elif choice == "5":
        print("Exited ")
        break

    else:
        print("Invalid choice ")
```

```python
# ATM System using Multiple Accounts

accounts = {
    1001: {"name": "Mohan", "pin": 1234, "balance": 15000},
    1002: {"name": "Netaji", "pin": 4321, "balance": 10000},
    1003: {"name": "AbhiRam", "pin": 1111, "balance": 20000}
}

def authenticate(acc_no, pin):
    if acc_no in accounts and accounts[acc_no]["pin"] == pin:
        return True
    return False

def check_balance(acc_no):
    print(f"Available Balance: ₹{accounts[acc_no]['balance']}")

def deposit(acc_no, amount):
    accounts[acc_no]["balance"] += amount
    print(f"₹{amount} deposited successfully ")

def withdraw(acc_no, amount):
    if amount <= accounts[acc_no]["balance"]:
        accounts[acc_no]["balance"] -= amount
        print(f"₹{amount} withdrawn successfully ")
    else:
        print("Insufficient balance ")


print("🏦 Welcome to ATM")

acc_no = int(input("Enter Account Number: "))
pin = int(input("Enter PIN: "))

if authenticate(acc_no, pin):
    print(f"\nLogin successful  Welcome {accounts[acc_no]['name']}")

    while True:
        print("\n1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            check_balance(acc_no)

        elif choice == "2":
            amt = int(input("Enter amount to deposit: "))
            deposit(acc_no, amt)

        elif choice == "3":
            amt = int(input("Enter amount to withdraw: "))
            withdraw(acc_no, amt)

        elif choice == "4":
            print("Thank you for using ATM ")
            break

        else:
            print("Invalid option ")

else:
    print("Invalid account number or PIN ")
```

```python
# Student Gradebook

students = {}

def add_student(name, marks):
    students[name] = marks
    print("Student added successfully")

def calculate_grade(average):
    if average >= 90:
        return "A+"
    elif average >= 80:
        return "A"
    elif average >= 70:
        return "B"
    elif average >= 60:
        return "C"
    else:
        return "Fail"

def view_report(name):
    if name not in students:
        print("Student not found")
        return

    marks = students[name]
    total = sum(marks)
    average = total / len(marks)
    grade = calculate_grade(average)

    print("\nReport Card")
    print("Name:", name)
    print("Marks:", marks)
    print("Total:", total)
    print("Average:", average)
    print("Grade:", grade)

# Main Menu
while True:
    print("\n1. Add Student")
    print("2. View Report")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        marks = list(map(int, input("Enter marks separated by space: ").split()))
        add_student(name, marks)

    elif choice == "2":
        name = input("Enter student name: ")
        view_report(name)

    elif choice == "3":
        print("Program ended")
        break

    else:
        print("Invalid choice")
```
